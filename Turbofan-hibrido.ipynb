```python
###RESOLUCION PARA h=0

h = 0 #h en metros
T0 = temp_isa(h)
T1 = T0 
p0 = pres_isa(h)
p1 = p0 
vel1 = 95 #Velocidad minima en nudos
vel2 = 501 #Velocidad maxima en nudos+1 para poder coger el 500
D_tr = 0.715 #m
A_tr = np.pi*(D_tr**2)/4 #Area interna del generador de gas
p8 = p0
rest_temp = 150
f_ini = 0.35
for v in range (vel1,vel2,15):
        
    v1 = v*1852/3600
    ro = p0/(Ra*T0)
    Mv = v1/(np.sqrt(gamma(T1)*Ra*T1))
    D_efectivo = 2*(1.70*D_fan/3.85) #Calculado en un apartado del informe.
    A_fan = np.pi*((D_efectivo**2)/4)
    G_fan = ro*A_fan*v1
    lista15.append ((f"{A_fan:.4f}"))
    
    
        ##DIFUSOR (di)    
    T1t = T1 + ((v1**2/(2*cp(T1))))
    p1 = p11 = p0 
    p1t = ((1+(((gamma(T1t)-1)*(Mv**2)))/2)**(gamma(T1t)/(gamma(T1t)-1)))*p1
    p2t = p1t
    T2t = T1t
    T12t = T1t #12 es la entrada al fan por la parte "secundaria", aunque aun no hay division fisica
    p12t = p2t #2 es la entrada al fan por la parte "primaria", aunque aun no hay division fisica
    M_di = np.sqrt(((T1t/T1)-1)*(2/(gamma(T1t)-1)))
    lista1.append (f"{v1:.2f}")
    lista2.append (f"{G_fan:.2f}")
    
   
    ##Fan (fan)
    G12 = G2 = G_fan
    G_pi = ro *A_tr* v1 #El correspondiente al turborreactor, se coge el area mayor 75x75 cm
    G_sig = G_fan - G_pi #sig corresponde a sigma y al gasto secundario
    lam = G_sig/G_pi
    A_di = G_pi*(lam+1)/(ro*v1) #Me sale que es igual que el area del fan
    G1 = ro*A_di*v1
        
    def  pi_fan(G):
        if 0 < G < 50:
            pi_fan = (1.5*G+50)/50
            return pi_fan
        if 50 < G < 100:
            pi_fan = (G-50+250)/100
            return pi_fan
        if 100 < G < 400:
            pi_fan = ((G-100)/300)*0.25 + 3 
            return pi_fan
        if G > 400:
            pi_fan = ((G-100)/300)*0.2 + 3 #De esta manera continua aumentando la presion con el gasto.
            return pi_fan
    
    def  re_fan(G):   
        if 0 < G < 50:
            re_fan = 0.75
            return re_fan
        if 50 < G < 100:
            re_fan = 0.85
            return re_fan
        if 100 < G < 400:
            re_fan = 0.72
            return re_fan 
        if G > 400:
            re_fan = 0.72
            return re_fan
            
            
    T25t_id = ((((((pi_fan(G_fan)))**((1004-1)/1004))-1)/(re_fan(G_fan)))+1)*T2t                                 #Se usa un T25t ideal para poder hacer la media
    T25t = (((((pi_fan(G_fan))**((gamma((T25t_id+T2t)/2)-1)/gamma((T25t_id+T2t)/2)))-1)/(re_fan(G_fan)))+1)*T2t
    p25t = p2t*(pi_fan(G_fan))
    G25 = G_pi 
    W_fan = G25*cp(T25t)*T25t - G_pi*cp(T2t)*T2t #Esto seria asi o no pq tenemos dos flujos 
    T13t = T25t
    p13t = p25t
    G13 = G_sig       
    tau_fan = cp((T13t+T12t)/2)*(T13t-T12t)       
    
        
    ##COMPRESOR DE BAJA (cb) 
    T26t_id = (((((pi_cb)**((1004-1)/1004))-1)/re_cb)+1)*T25t #Se usa un T25t ideal para poder hacer la media
    T26t = (((((pi_cb)**((gamma((T26t_id+T25t)/2)-1)/gamma((T26t_id+T25t)/2)))-1)/re_cb)+1)*T25t
    p26t = p25t*pi_cb
    G26 = G25 
    W_cb = G26*cp(T26t)*T26t - G25*cp(T25t)*T25t
    
             
    ##JET-PIPE1 (jp_1)
    G27 = G26 - Gs
    T27t = T26t
    p27t = 0.98*p26t
    
                
    ##COMPRESOR DE ALTA ELECTRICO(ca)
    re_ca = 0.95 - (0.2*Mv) 
    pi_ca = 2 + 5*Mv 
    T3t_id = (((((pi_ca)**((1004-1)/1004))-1)/re_ca)+1)*T27t #Se usa un T3t ideal para poder hacer la media
    T3t = (((((pi_ca)**((gamma((T3t_id+T27t)/2)-1)/gamma((T3t_id+T27t)/2)))-1)/re_ca)+1)*T27t 
    p3t = p27t*pi_ca
    G3 = G27 
    W_ca = G3*cp(T3t)*T3t - G27*cp(T27t)*T27t #Es el trabajo que ha de proporcionr el motor eléctrico
    
    lista30.append (f"{pi_ca:.2f}")
    
        
    ##CAMARA DE COMBUSTION
    p4t = p3t*0.95
    fs = Mc/(4.76*(Mair)*12.5)
    f = f_ini*fs
    c = f*G3
    G4 = G3 + c
    T4t_id = ((re_cc*c*L) + (G3*1004*T3t))/(G4*1004)
    T4t = ((re_cc * c* L) + (G3 * cp((T4t_id+T3t)/2) * T3t))/(G4*cp((T4t_id+T3t)/2))
         
        
    ##TURBINA (t) 
    G5 = G4
    W_t = -(W_fan+W_cb)

    T5t_id = (W_t + G4*cp(1004)*T4t)/(G5*cp(1004))
    T5t = (W_t + G4*cp(T4t)*T4t)/(G5*cp(T5t_id))
    p5t = (((((T5t/T4t)-1)/re_t)+1)**(gamma((T5t+T4t)/2)/(gamma((T5t+T4t)/2)-1)))*p4t 
        
    T5t_f = T5t - rest_temp #Temperatura de salida de gases reducida
    W_t_f = G5*cp(T5t_f)*T5t_f - G4*cp(T4t)*T4t #seia el nuevo trabajo del eje.
            
    dif_w = (W_t_f*(-1)) - (W_fan+W_cb) #Extra de trabajo que saca la turbina
    
    prop_W_fan = (W_fan/-W_t) #Relacion que aporta el fan al trabajo de la turbina inicial. 
    prop_W_cb = (W_cb/-W_t) #Relacion que aporta el compresor de baja al trabajo de la turbina inicial.
    
    difw_fan = dif_w * prop_W_fan #Extra de trabajo que llega al fan 
    difw_cb = dif_w * prop_W_cb #Extra de trabajo que llega al comresor de baja
    
    W_fan2 = difw_fan + W_fan
    W_cb2 = difw_cb + W_cb
    
    
    ##FAN2
    T25t_f_id = (W_fan2 + G_pi*cp(1004)*T25t)/(G25*cp(1004))
    T25t_f = (W_fan2 + G_pi*cp( T25t_f_id)*T2t)/(G25*cp(T25t_f_id))
    T13t_f = T25t_f
    
    pi_fan2 = ((re_fan(G_fan)*((T25t_f/T2t)-1)) + 1)**(gamma((T25t_f+T2t)/2)/(gamma((T25t_f+T2t)/2)-1))
    pi_fan1 = ((re_fan(G_fan)*((T25t/T2t)-1)) + 1)**(gamma((T25t+T2t)/2)/(gamma((T25t+T2t)/2)-1))
          
    varT25t = T25t_f-T25t    
    p25t_f = p2t*(pi_fan2)
    p13t_f = p25t_f
    G13 = G_sig       
    tau_fan = cp((T13t_f+T12t)/2)*(T13t_f-T12t)
    
    lista37.append (f"{pi_fan2:.2f}")
    
    
    ##COMPRESOR DE BAJA2 (cb)
    T26t_f_id = (W_cb2 + G_pi*cp(T25t_f)*T25t_f)/(G25*cp(1004))
    T26t_f = (W_cb2 + G_pi*cp(T25t_f)*T25t_f)/(G25*cp(T26t_id)) 
    varT26t = T26t_f-T26t
    pi_cb2 = ((re_cb*((T26t_f/T25t)-1)) + 1)**(gamma((T26t_f+T25t)/2)/(gamma((T26t_f+T25t)/2)-1)) 
    p26t_f = p25t_f*(pi_cb2)
    G26 = G25     
   
    lista38.append (f"{pi_cb2:.2f}")
    

    ##JET-PIPE1 (jp_1)
    G27 = G26 - Gs
    T27t_f = T26t_f
    p27t_f = 0.98*p26t_f
    
                
    ##COMPRESOR DE ALTA ELECTRICO(ca)
    re_ca = 0.95 - (0.2*Mv) 
    pi_ca = 2 + 5*Mv 
    T3t_id_f = (((((pi_ca)**((1004-1)/1004))-1)/re_ca)+1)*T27t_f #Se usa un T3t ideal para poder hacer la media
    T3t_f = (((((pi_ca)**((gamma((T3t_id_f+T27t_f)/2)-1)/gamma((T3t_id_f+T27t_f)/2)))-1)/re_ca)+1)*T27t_f
    p3t_f = p27t_f*pi_ca
    G3 = G27 
    W_ca = G3*cp(T3t_f)*T3t_f - G27*cp(T27t_f)*T27t_f #Es el trabajo que ha de proporcionr el motor eléctrico
    
        
    ##CAMARA DE COMBUSTION
    p4t_f = p3t_f*0.95
    fs = Mc/(4.76*(Mair)*12.5)
    f = f_ini*fs
    c = f*G3
    G4 = G3 + c
    T4t_id_f = ((re_cc*c*L) + (G3*1004*T3t_f))/(G4*1004)
    T4t_f = ((re_cc * c* L) + (G3 * cp((T4t_id+T3t)/2) * T3t_f))/(G4*cp((T4t_id_f+T3t_f)/2))
    
         
    ##TURBINA2 (t) 
    G5 = G4
    W_t2 = -(W_fan2+W_cb2)
    T5t_f2_id = (W_t2 + G4*cp(1004)*T4t_f)/(G5*cp(1004))            
    T5t_f2 = (W_t2 + G4*cp(T4t_f)*T4t_f)/(G5*cp(T5t_f2_id))
    p5t_f = (((((T5t_f2/T4t_f)-1)/re_t)+1)**(gamma((T5t_f2+T4t_f)/2)/(gamma((T5t_f2+T4t_f)/2)-1)))*p4t_f
    
    lista3.append (f"{G_pi:.2f}")
    lista4.append (f"{G_sig:.2f}")
    lista5.append (f"{lam:.4f}")
    lista6.append (f"{A_di:.6f}")
    lista7.append (f"{pi_fan(G_fan):.4f}")
    lista8.append (f"{re_fan(G_fan):.4f}") 
    lista14.append (f"{T25t:.2f}")
    lista32.append (pi_cb)
    lista21.append (f"{T5t:.2f}")
    lista22.append (f"{T5t_f2:.2f}")
    lista23.append (f"{W_t2:.2f}")
    lista29.append (f"{T4t_f:.2f}")       
    
    
    ##TOBERA PRIMARIA (top) #se busca que sea crítica, es decir mach de salida = 1

    G7 = G5
    G8 = G7        
    T7t = T5t_f2
    T8t = T7t
    p7t = p5t_f
    p8t = p7t  
    
    T8 = T8t*(2/(gamma(T8t)+1)) #Formula para M=1 
    v8 = np.sqrt((T8t-T8)*2*(((cp(T8t)+cp(T8))/2)))
    p8 = (((((T8/T8t)-1)/re_top)+1)**(gamma(T8t)/(gamma(T8t)-1)))*p8t 
    ro8 = p8/(Ra*T8)
    A8 = G8/(ro8*v8)
    
    M_top = np.sqrt(((T8t/T8)-1)*(2/(gamma(T8t)-1)))         
    E_top = G8*v8 - G_pi*v1 + A8*(p8-p1)
    D_top = np.sqrt(A8*4/np.pi) #Diametro (interno) del generador de gas   
    
    if T8t > 2950:
        lista9.append ("Ex")
    if T8t < 2950:
        lista9.append (f"{T8t:.2f}")
      
    lista10.append (f"{A8:.6f}")
    lista11.append (f"{M_top:.1f}") 
    lista12.append (f"{E_top:.1f}")  
    lista27.append (f"{p8t:.2f}")
    lista28.append (f"{p8:.2f}")
    lista33.append (f"{T8:.2f}")
        
    
    ##TOBERA SECUNDARIA (tos) Se busca Mach critico
    G17 = G13 
    G18 = G17
    T17t= T13t_f
    T18t = T17t
    p17t = p13t_f 
    p18t = p17t 

    T18 = T18t*(2/(gamma(T18t)+1)) #Formula para M=1
    p18 = (((((T18/T18t)-1)/re_tos)+1)**(gamma(T18t)/(gamma(T18t)-1)))*p18t   
    v18 = np.sqrt((T18t-T18)*2*cp(T18t))
    ro18 = p18/(Ra*T18)
    A18 = G18/(ro18*v18)
    M_tos = np.sqrt(((T18t/T18)-1)*(2/(gamma(T18t)-1)))         
    E_tos = G18*v18 - G_sig*v1 + A18*(p18-p11)
        
    if T18t > 1950:
        lista16.append ("Ex")
    if T18t < 1950:
        lista16.append (f"{T18t:.2f}")   
        
    cs = (c)/(E_tos+E_top) #Consumo especifico   
     
    re8_motor = ((G8)*(v8**2) - G_pi*(v1**2))/(2*c*L)
    re8_prop = (2*v1*((G8)*v8 - G_pi*v1))/((G8)*(v8**2) - G_pi*(v1**2))
    re8_motor_prop = (v1*((G8)*v8 - G_pi*v1))/(c*L)
    
    re18_motor = ((G18)*(v18**2) - G_sig*(v1**2))/(2*(c)*L)
    re18_prop = (2*v1*((G18)*v18 - G_sig*v1))/((G18)*(v18**2) - G_sig*(v1**2))
    re18_motor_prop = (v1*((G18)*v18 - (G_sig*v1)))/((c)*L)
    
    re_motor = (G_pi/G_fan)*re8_motor + (G_sig/G_fan)*re18_motor 
    re_prop = (G_pi/G_fan)*re8_prop + (G_sig/G_fan)*re18_prop
    re_motor_prop = (G_pi/G_fan)*re8_motor_prop + (G_sig/G_fan)*re18_motor_prop
   
    E = E_top+E_tos 
    
    lista13.append (f"{E:.1f}")
    lista18.append (f"{M_tos:.1f}")   
    lista17.append (f"{A18:.6f}") 
    lista19.append (f"{E_tos:.1f}")   
    lista20.append (f"{cs:.8f}")
    lista24.append (f"{re_motor:.3f}")
    lista25.append (f"{re_prop:.3f}")
    lista26.append (f"{re_motor_prop:.3f}")
    lista31.append (f"{G18:.1f}")
    lista34.append (f"{T18:.2f}")
    lista35.append (f"{p18:.2f}")
    lista36.append (f"{p18t:.2f}")
    
    
    diccionario = {"velocidad (m/s)" :  lista1 , "M_top" : lista11 , "M_tos" : lista18 ,
                   "E_top (N)" : lista12 , "E_tos (N)" : lista19 , "E (N)" : lista13 , 
                   "cs" : lista20 , "G_fan (Kg/s)" : lista2 , "G_pi (Kg/s)" : lista3 , 
                   "G_sig (Kg/s)" : lista4 , "Lambda" : lista5 , "A_di (m^2)" : lista6 ,
                   "Pi_fan" : lista7 , "pi_cb" : lista32 , "re_fan" : lista8 , 
                   "T8t (K)" : lista9 , "A8 (m^2)" : lista10 , "p8t (pa)" : lista27 , 
                   "p8 (pa)" : lista28 , "T18t (K)" : lista16 , "A18 (m^2)" : lista17 ,
                   "re_motor" : lista24 , "re_prop" : lista25 , "re_motor_prop" : lista26 , 
                   "T25t (K)" : lista14 , "T4t (K)" : lista29 , "T5t (K)" : lista21 ,
                   "T5t_f2 (K)" : lista22 , "W_t_f (W)" : lista23}

       
    df = pd.DataFrame.from_dict(diccionario)    
```
